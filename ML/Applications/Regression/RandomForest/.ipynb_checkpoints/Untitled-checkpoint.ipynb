# Ensemble learning

Ensemble Learning is a technique where multiple models (often called "weak learners") are trained and combined to solve the same problem. The idea is that a group of models working together can outperform a single model.

### Working

1. Bootstrap Sampling
Create multiple random subsets of the original dataset with replacement. Each subset is used to train a different decision tree.This introduces diversity among trees.

2. Train Multiple Decision Trees
Each tree is trained independently and in parallel.At each split in a tree:
    Only a random subset of features is considered (not all).
    This prevents trees from becoming too similar (reduces correlation).

3. Make Predictions
Once all trees are trained, predictions are made:
    Classification: each tree votes for a class label. The majority vote is the final prediction.
    Regression: each tree predicts a value. The average of all predictions is the final result.

4. Final Output (Aggregation)
The predictions from all trees are aggregated to make the final prediction. This reduces overfitting, improves generalization, and stabilizes the model.
